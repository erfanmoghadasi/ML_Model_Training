```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, 0.1)
y = 2*x + 3
y_noise = np.random.normal(size=x.size)
y_data = y + y_noise
plt.plot(x, y)
plt.scatter(x, y_data, color='red', s=5)
plt.xlabel('Dependent variable')
plt.ylabel('Independent variable')
plt.legend(['Original data', 'Noisy data'])
plt.show()
```

```python
x = np.arange(-5, 5, 0.1)
y = x**3 + x**2 + x
y_noise = np.random.normal(size=x.size) * 10
y_data = y + y_noise
plt.plot(x, y)
plt.scatter(x, y_data, color='red', s=5)
plt.xlabel('Dependent variable')
plt.ylabel('Independent variable')
plt.legend(['Original data', 'Noisy data'])
plt.show()
```

### Quadratic
$$
Y = X^2
$$

```python
x = np.arange(-5, 5, 0.1)
y = x**2
y_noise = np.random.normal(size=x.size) * 2
y_data = y + y_noise
plt.plot(x, y)
plt.scatter(x, y_data, color='red', s=5)
plt.xlabel('Dependent variable')
plt.ylabel('Independent variable')
plt.legend(['Original data', 'Noisy data'])
plt.show()
```

### Exponential
$$
Y = a + bc^x
$$

```python
x = np.arange(-5, 5, 0.1)
# y = 1 + 2 * (3 ** x)
y = np.exp(x)
y_noise = np.random.normal(size=x.size) * 5
y_data = y + y_noise
plt.plot(x, y)
plt.scatter(x, y_data, color='red', s=5)
plt.xlabel('Dependent variable')
plt.ylabel('Independent variable')
plt.legend(['Original data', 'Noisy data'])
plt.show()
```

### Logarithmic
$$
Y = log(X)
$$

```python
x = np.arange(-5, 5, 0.1)
y = np.log(x)
plt.plot(x, y)
plt.xlabel('Dependent variable')
plt.ylabel('Independent variable')
plt.show()
```

### Sigmoidal / Logistic
$$
Y = a + \frac{b}{1 + c^(x - d)}
$$

```python
x = np.arange(-5, 5, 0.1)
y = 1 - (4 / (1 + np.power(3, x-2)))

plt.plot(x, y)
plt.xlabel('Dependent variable')
plt.ylabel('Independent variable')
plt.show()
```
